## Personalizando Qiskit.

En  Qiskit puedo personalizar el output de cómo el circuito se renderiza y la información que renderizar

```python
from qiskit import *
from matplotlib.pyplot import plot as plt
import numpy as np
```

```python
circuit = QuantumCircuit(5,5)
```

```python
for q in range(0,5):
    circuit.h(q)
    circuit.measure(q,q)
```

```python
circuit = QuantumCircuit(5,5)
```

```python
# fold : es un parámetro útil para cuando tenemos circuitos muy largosy queremos imprimirlo
# Por defecto se muestran hasta 25 compeurtas del circuito
# Primero crearemos un circuito
```

```python
def create_big_circuit(circuit):
    for q in range(0,5):
        circuit.h(q)
        circuit.x(q)
        circuit.y(q)
        circuit.z(q)
        circuit.h(q)

    circuit.barrier()

    for q in range(0,5):
        circuit.h(q)
        if q < 4:
            circuit.cx(q,q+1)
            circuit.cy(q,q+1)
            circuit.cz(q,q+1)
        circuit.h(q)

    circuit.barrier()

    for q in range(0,5):
        circuit.h(q)
        circuit.s(q)
        circuit.t(q)
        circuit.h(q)
    return circuit
```

```python
circuit = create_big_circuit(circuit)
```

```python
circuit.draw(output='mpl',fold=True) # Ajusta la impresión para que se vea todo.
```

```python
circuit.draw(output='mpl',fold=20) # Puedo decirle le número de compeurtas que quier ver, es decir de a cuanto em meustra
```

```python
circuit.draw("mpl", style="bw", fold=-1)
```

## Accesibilidad

Podemos cambiar el fondo, formatos y tamaño de fuentes.

```python
options ={
    "fontsize": 20,
    "subfontsize": 25,  
}

circuit.draw("mpl", style=options, fold=-1)
```

En algunas ocasiones si quieramos tener unmrbes personalizados del circuito:


```python
gate_names = {
    "displaytext": {
        # gate name : gate text to be displayed
        "x": "Pauli X",
        "y": "Pauli Y",
        "z": "Pauli Z",
        "h": "Hadamard",
        "textcolor": "#6fdc8c",
        "linecolor": "#da1e28",
        "background_color":"black"

    },
}


```

```python
circuit.draw(output='mpl',style=gate_names, fold=-1, interactive=True)

```
